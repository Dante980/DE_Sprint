```python
import time # Импортируем библиотеку time
from IPython import display # Импортируем библиотеку IPython
for k in range (-50,50): # Коэфициент k
    x = np.arange (-10,10) # Диапазон x
    y = np.arange (-10,10) * k # Диапазон y только с множителем k
    plt.plot(x,y) # Вывод графика
    display.display(plt.gcf()) # Выводится как дисплей
    display.clear_output(wait=True) # Удаление ненужного, что ранее вывелось
    time.sleep(1) # Запускать графика
```

# Датасет

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
```

```python
with open("./dataset.txt", mode="r") as file:
    for human in file.readlines():
        age.append(int(human.split(",")[0]))
        
        if human.split(",")[1] == "male":
         sex.append(1)
        else:
          sex.append(0)
        bmi.append(float(human.split(",")[2]))
        
        children.append(float(human.split(",")[3]))
        
        if human.split(",")[4] == "yes":
          smoker.append(1)
        else:
          smoker.append(0)
        if human.split(",")[5] == "northwest":
          region.append(0)
        elif human.split(",")[5] == "northeast":
          region.append(1)
        elif human.split(",")[5] == "southwest":
          region.append(2)
        elif human.split(",")[5] == "southeast":
          region.append(3)
        else:
          region.appen(human.split(",")[5]) 
        charges.append(float(human.split(",")[6]))
```
