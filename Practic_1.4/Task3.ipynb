```python
import matplotlib.pyplot as plt
names = ["year", "sex", "some_digit_1", "single_digit", "alive", "stream", "salary"]
#      csv_data
csv_data = {"year":[], "sex":[], "some_digit_1":[], "single_digit":[], "alive":[], "stream":[], "salary":[]}
#           
with open("data.csv", "r") as file:
    #       
    for line in file:
        #     
        temp = line.split(",")
        #    
        csv_data["year"].append(int(temp[0]))

        if temp[1] == "male":
            csv_data["sex"].append(1)
        else:
            csv_data["sex"].append(0)
        csv_data["some_digit_1"].append(float(temp[2]))
        csv_data["single_digit"].append(int(temp[3]))
        if temp[4] == "no":
            csv_data["alive"].append(0)
        else:
            csv_data["alive"].append(1)
        csv_data["stream"].append(temp[5])
        csv_data["salary"].append(float(temp[6].replace("\n", "")))
        print(temp)
# print(csv_data["year"][1], csv_data["sex"][1], csv_data["some_digit_1"][1], csv_data["single_digit"][1], csv_data["stream"][1], csv_data["magnitude"][1])   
```
