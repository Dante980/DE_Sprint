```python
import matplotlib.pyplot as plt
# поключаем библиотеку для рисования
import numpy as np
# поключаем библиотеку для вычиления

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
# задаем два массва. Запоняем его числами (100 штук) от -2 до 2 

X, Y = np.meshgrid(x, y)
# создаем матрицу, сораную из координат x и y. 
# Проще говоря в X будут все возможные x координаты
# А в Y будут все возможные координаты y

z = (X**2 + Y**2 - 1)**3 - X**2 * Y**3
# строим нашу фунцию

plt.contour(X, Y, z, [0])
# строим контур в xy диапозоне по функции z, где уровень (значение) равняется 0 
```
